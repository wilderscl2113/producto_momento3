SE FILTRA LOS DATOS PARA QUE SOLO SE MUESTREN LOS QUE TIENEN EN GRUPO DE CLIENTES LA "CATEGORIA" A

```python
filtroA = data['Grupo de clientes']== 'A'
filtroA
dataFilterA= data[filtroA]
dataFilterA
```

SE FILTRA LOS DATOS PARA QUE SOLO SE MUESTREN LOS QUE TIENEN EN GRUPO DE CLIENTES LA "CATEGORIA" B

```python
filtroB = data['Grupo de clientes']== 'B'
filtroB
dataFilterB= data[filtroB]
dataFilterB
```

SE FILTRA LOS DATOS PARA QUE SOLO SE MUESTREN LOS QUE TIENEN EN GRUPO DE CLIENTES LA "CATEGORIA" E

```python
filtroE = data['Grupo de clientes']== 'E'
filtroE
dataFilterE= data[filtroE]
dataFilterE
```

SE HACE LA CUENTA DE CUANTOS CAMPOS HAY DE CADA UNA DE LAS CATEGORIAS DE LOS CLIENTE ESTO CON EL FIN DE SABER SI HAY DIFERENCIAS Y TENER LA REFERENCIA DE LA CANTIDAD DE DATMOS QUE ESTE TIENE

```python
countA=dataFilterA['Grupo de clientes'].count()
countB=dataFilterB['Grupo de clientes'].count()
countE=dataFilterE['Grupo de clientes'].count()

countA,countB,countE
```

SE CONCATENAN LOS RESULTADOS FILTRADOS Y SE CREA UN SOLO DF

```python
result = pd.concat([dataFilterA,dataFilterB,dataFilterE],ignore_index=True)
```

```python
result
```

SE CUENTA EN EL NUEVO DF CREADO LA CANTIDAD DE VALORES DE CADA UNO DE LOS DATOS DE LA COLUMNA GRUPO DE CLIENTES

```python
conteo_grupos = result['Grupo de clientes'].value_counts()
```

SE LE DA UN TAMAÑO A LA FIGURA , SE TOMA EL NUMERO DE CONTEO DE LOS DATOS, SU RESPECTI VO GRUPO Y SE LE DA COLOR , SE AGREGA EL TITULO GENERAL, EL TITULO PARA EL EJE X Y EL EJE Y Y SE MUESTRA

```python
colores = ['skyblue', 'salmon', 'lightgreen']
plt.figure(figsize=(10, 6))
plt.bar(conteo_grupos.index, conteo_grupos, color=colores)
plt.title('Número de Clientes por Grupo')
plt.xlabel('Grupo de Clientes')
plt.ylabel('Número de Clientes')
plt.show()
```
